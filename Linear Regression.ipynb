## Model Selection

```python
X = df[df["data"] == "Train"]
Y = df[df["data"] == "Test"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
ld_new_train, ld_val = train_test_split(X, test_size=0.2, random_state=100)
```

```python
train_features = ld_new_train.drop(['Interest.Rate','data'],axis=1) ## this contains features for train

train_targets = ld_new_train['Interest.Rate'] ### this contains the actual value for train set

val_features = ld_val.drop(['Interest.Rate','data'],axis=1) ### this contains features for validation set

val_targets = ld_val['Interest.Rate'] 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
cols=["Amount.Requested","Loan.Length","Debt.To.Income.Ratio","Monthly.Income","FICO.Range","Open.CREDIT.Lines","Revolving.CREDIT.Balance","Inquiries.in.the.Last.6.Months","Employment.Length"]
```

```python
X_test=Y
```

```python
for i in cols:
    
    # fit on training data column
    scale = StandardScaler().fit(train_features[[i]])
    
    # transform the training data column
    train_features[i] = scale.transform (train_features[[i]])
    val_features[i] = scale.transform (val_features[[i]])
    
    # transform the testing data column
    X_test[i] = scale.transform(X_test[[i]])
```
